### Data Preprocessing

#### Split DataFrame in train and test

```python
# input variable
X = cancer_df.drop(['target'], axis = 1)
X.shape
```

```python
# output variable
y = cancer_df['target']
y.shape
```

```python
# split dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 5)
```

<br>
<br>

### Feature Scaling

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

<br>
<br>
<br>

## Breast Cancer Detection Machine Learning Model Building

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

<br>
<br>

### Logistic Regression

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression(random_state = 51)
lr_classifier.fit(X_train, y_train)
y_pred_lr = lr_classifier.predict(X_test)
log_acc = accuracy_score(y_test, y_pred_lr)
log_acc
```

```python
# Train with Standard scaled Data
lr_classifier2 = LogisticRegression(random_state = 51)
lr_classifier2.fit(X_train_sc, y_train)
y_pred_lr_sc = lr_classifier.predict(X_test_sc)
accuracy_score(y_test, y_pred_lr_sc)
```

```python

logreg_cm = confusion_matrix(y_test, y_pred_lr)
f, ax = plt.subplots(figsize=(5,5))
cmap = sns.light_palette("#3fdd01", as_cmap=True)
sns.heatmap(logreg_cm, annot=True, linewidth=0.7, linecolor='red', fmt='g', ax=ax, cmap=cmap)
plt.title('Logistic Regression Classification Confusion Matrix')
plt.xlabel('Y predict')
plt.ylabel('Y test')
plt.show()
```

<br>
<br>

### Decision Tree Classifier

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 51)
dt_classifier.fit(X_train, y_train)
y_pred_dt = dt_classifier.predict(X_test)
dec_acc = accuracy_score(y_test, y_pred_dt)
dec_acc
```

```python
# Train with Standard scaled Data
dt_classifier2 = DecisionTreeClassifier(criterion = 'entropy', random_state = 51)
dt_classifier2.fit(X_train_sc, y_train)
y_pred_dt_sc = dt_classifier.predict(X_test_sc)
accuracy_score(y_test, y_pred_dt_sc)
```

```python
# The confusion matrix
dtcla_cm = confusion_matrix(y_test,y_pred_dt )
f, ax = plt.subplots(figsize=(5,5))
cmap = sns.light_palette("#3fdd01", as_cmap=True)
sns.heatmap(dtcla_cm, annot=True, linewidth=0.7, linecolor='red', fmt='g', ax=ax, cmap=cmap)
plt.title('Decision Tree Classification Confusion Matrix')
plt.xlabel('Y predict')
plt.ylabel('Y test')
plt.show()
```

<br>
<br>

### Support Vector Classifier

```python
# Support vector classifier
from sklearn.svm import SVC
svc_classifier = SVC()
svc_classifier.fit(X_train, y_train)
y_pred_scv = svc_classifier.predict(X_test)
sv_acc = accuracy_score(y_test, y_pred_scv)
sv_acc
```

```python
# Train with Standard scaled Data
svc_classifier2 = SVC()
svc_classifier2.fit(X_train_sc, y_train)
y_pred_svc_sc = svc_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_svc_sc)
```

```python
# The confusion matrix
svmcla_cm = confusion_matrix(y_test,y_pred_scv )
f, ax = plt.subplots(figsize=(5,5))
cmap = sns.light_palette("#3fdd01", as_cmap=True)
sns.heatmap(svmcla_cm, annot=True, linewidth=0.7, linecolor='red', fmt='g', ax=ax, cmap=cmap)
plt.title('SVM Classification Confusion Matrix')
plt.xlabel('Y predict')
plt.ylabel('Y test')
plt.show()
```

<br>
<br>

### Random Forest Classifier

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf_classifier.fit(X_train, y_train)
y_pred_rf = rf_classifier.predict(X_test)
rf_acc = accuracy_score(y_test, y_pred_rf)
rf_acc
```

```python
# Train with Standard scaled Data
rf_classifier2 = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf_classifier2.fit(X_train_sc, y_train)
y_pred_rf_sc = rf_classifier.predict(X_test_sc)
accuracy_score(y_test, y_pred_rf_sc)
```

```python
# The confusion matrix
rfcla_cm = confusion_matrix(y_test,y_pred_rf)
f, ax = plt.subplots(figsize=(5,5))
cmap = sns.light_palette("#3fdd01", as_cmap=True)
sns.heatmap(rfcla_cm, annot=True, linewidth=0.7, linecolor='red', fmt='g', ax=ax, cmap=cmap)
plt.title('Random Forest Classification Confusion Matrix')
plt.xlabel('Y predict')
plt.ylabel('Y test')
plt.show()
```
